# Activation Function

An activation function is a function that is added to an **artificial neural network** in order to help the network learn complex patterns in the data. When comparing with a neuron-based model that is in our brains, the activation function is at the end deciding what is to be fired to the next neuron.

![neron network mechanism](https://miro.medium.com/v2/resize:fit:720/format:webp/1*0wDAsK-gy8NMxMyimPjw_g.png)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

### Types of Activation Functions

#### 1. Sigmoid Function

The Sigmoid Function looks like an S-shaped curve. $f(z) = \dfrac{1}{(1+ e^{-z})}$

```python
#define the Sigmoid Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

The Sigmoid Function looks like an **S-shaped** curve.

```python
# 生成x值
x = np.linspace(-10, 10, 400)
# 计算y值
y = sigmoid(x)
# 绘图
plt.plot(x, y)
# 添加网格
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
# 轴中心线
plt.axhline(y=0.5, color='k', linewidth=0.5) # 添加水平X=0轴线
plt.axvline(x=0, color='k', linewidth=0.5) # 添加垂直Y=0轴线
# 标题和标签
plt.title("Function plot of Sigmoid Function")
plt.xlabel("x")
plt.ylabel("y")
# 显示图像
plt.show()
```

**Advantages of Sigmoid Function**:

- The output of a sigmoid function ranges between 0 and 1. Since, output values bound between 0 and 1, it normalizes the output of each neuron.

- Specially used for models where predicting the probability as an output. Since the probability of anything exists only between the range of 0 and 1.

- Smooth gradient, preventing derivative discontinuity "jumps" in output values.

- The function is differentiable.That means, we can find the slope of the sigmoid curve at any two points.

- Clear predictions, since most of cases for the value are very close to 1 or 0 instead of being close to 0.5.

**Disadvantages of Sigmoid Function**:

- Prone to **gradient vanishing** (when the sigmoid function value is either too high or too low, the derivative becomes very small i.e. << 1. This causes vanishing gradients and poor learning for deep networks.)

  - The Sigmoid function maps any input value to a range between 0 and 1. As the input value becomes very large or very small, the Sigmoid function's output approaches 1 or 0, respectively. In these regions, the curve of the Sigmoid function flattens out, which means its derivative (or slope) approaches zero. During backpropagation in neural networks, gradients are calculated using these derivatives to update the weights. When the derivative is close to zero, the gradient becomes very small, leading to the "vanishing gradient" problem. As a result, neurons in the earlier layers of deep networks receive very small gradients, making their weights almost unchanged. This severely hampers the learning process, especially in deep neural networks, as the layers closer to the input do not learn effectively. 
  - Sigmoid 函数将任何输入值映射到 0 到 1 之间的范围。当输入值变得非常大或非常小时，Sigmoid 函数的输出趋向于 1 或 0。在这些区域，Sigmoid 函数的曲线变平缓，这意味着其导数（或斜率）接近于零。在神经网络中，通过这些导数来计算梯度，并用它们来更新权重。当导数接近零时，梯度变得非常小，导致"梯度消失"问题。结果是，深度网络中较早层的神经元接收到的梯度非常小，使得它们的权重几乎不发生变化。这严重阻碍了学习过程，尤其是在深度神经网络中，因为靠近输入的层效果学习不足。

- The function output is not centered on 0, which will reduce the efficiency of weight update.

  - The Sigmoid function outputs values in the range (0, 1). This means that the output is always positive and, on average, will be greater than 0.5 when inputs are randomly distributed around 0. In neural networks, this non-zero-centered output can lead to undesirable dynamics during weight updates. Specifically, if the output is consistently positive (or consistently negative for other activation functions), the gradients for the weights will also all be of the same sign, leading to either all weights being pushed to increase or all to decrease during the update. This can make the optimization less efficient and can lead to convergence to suboptimal solutions. Centering the output around 0 (as seen in functions like the hyperbolic tangent function) can help in more balanced and efficient learning.
  -  Sigmoid 函数的输出值在（0, 1）范围内。这意味着输出总是正的，并且当输入随机分布在 0 周围时，平均值会大于 0.5。在神经网络中，这种非零中心的输出可能会在权重更新期间导致不希望的动态变化。具体来说，如果输出始终为正（或对于其他激活函数始终为负），则权重的梯度也都将是同一符号，导致所有权重在更新期间都被推动增加或都减少。这会使优化效率变低，并可能导致收敛到次优解。以 0 为中心的输出（如双曲正切函数所见）可以帮助实现更平衡和高效的学习。

- The sigmoid function performs exponential operations, which is slower for computers.
  
  - The Sigmoid function is mathematically defined as, which involves an exponential operation. Exponential functions are computationally more intensive than simpler operations like addition, subtraction, or multiplication. This computational intensity is due to the series expansion used to calculate exponentials, which can slow down the training of neural networks, especially when dealing with large datasets or complex models. In contrast, other activation functions like ReLU (Rectified Linear Unit) are computationally more efficient because they involve simpler mathematical operations that are faster to compute. 
  - Sigmoid 函数数学上涉及到一个指数运算。指数函数在计算上比加、减、乘等简单运算更为密集。这种计算强度是由于用来计算指数的级数展开，这可能会减慢神经网络的训练速度，特别是在处理大型数据集或复杂模型时。相比之下，其他激活函数如 ReLU（修正线性单元）在计算上更为高效，因为它们涉及更简单的数学操作，计算速度更快。

#### 2. Tanh or Hyperbolic Tangent Activation Function

```python
#define the Sigmoid Function
def Tanh(x):
    return np.tanh(x)
```

```python
# 生成x值
x = np.linspace(-10, 10, 400)
# 计算y值
y = Tanh(x)
# 绘图
plt.plot(x, y)
# 添加网格
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
# 轴中心线
plt.axhline(y=0, color='k', linewidth=0.5) # 添加水平X=0轴线
plt.axvline(x=0, color='k', linewidth=0.5) # 添加垂直Y=0轴线
# 标题和标签
plt.title("Function plot of Sigmoid Function")
plt.xlabel("x")
plt.ylabel("y")
# 显示图像
plt.show()
```

The tanh activation function is also sort of sigmoidal (S-shaped).     $f(x) = \tanh(x) = \frac{2}{1+e^{-2x}} - 1$

#### Comparison between Sigmoid and Tanh Activation Functions

- First of all, when the input is large or small, the output is almost smooth and the gradient is small, which is not conducive to weight update. The difference is the output interval. The output interval of tanh is 1, and the whole function is 0-centric, which is better than sigmoid.
- 首先，当输入较大或较小时，输出几乎是平滑的，梯度较小，不利于权重更新。区别在于输出区间。tanh 的输出区间为 1，整个函数以 0 为中心，比 sigmoid 好。
- The major advantage is that the negative inputs will be mapped strongly negative and the zero inputs will be mapped near zero in the tanh graph.
- 其主要优势在于，负输入将被映射为强负值，而零输入将被映射为接近 tanh 图形中的零值。


***Note:*** In general binary classification problems, the tanh function is used for the hidden layer and the sigmoid function is used for the output layer. However, these are not static, and the specific activation function to be used must be analyzed according to the specific problem, or it depends on debugging.

### 3. ReLU (Rectified Linear Unit) Activation Function

```python
def ReLU(x):
    return np.maximum(0, x)
```

```python
# 生成x值
x = np.linspace(-10, 10, 400)
# 计算y值
y = ReLU(x)
# 绘图
plt.plot(x, y)
# 添加网格
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
# 轴中心线
plt.axvline(x=0, color='k', linewidth=0.5) # 添加垂直Y=0轴线
# 标题和标签
plt.title("Function plot of Relu Function")
plt.xlabel("x")
plt.ylabel("y")
# 显示图像
plt.show()
```

The ReLU is half rectified (from the bottom). f(z) is zero when z is less than zero and f(z) is equal to z when z is above or equal to zero.

$$\sigma(x) = 
\begin{cases} 
\max(0, x) & ,\ x \geq 0 \\
0 & ,\ x < 0 
\end{cases}$$

The ReLU (Rectified Linear Unit) function is an activation function that is currently more popular compared to other activation functions in deep learning.

**Advantages of Sigmoid Function**:

Compared with the sigmoid function and the tanh function, it has the following advantages:

When the input is positive, there is no gradient saturation problem.
The calculation speed is much faster. The ReLU function has only a linear relationship. Whether it is forward or backward, it is much faster than sigmoid and tanh. (Sigmoid and tanh need to calculate the exponent, which will be slower.)

**Disadvantages of Sigmoid Function**:

Of course, there are disadvantages:

1) Dead ReLU problem- When the input is negative, ReLU is completely inactive, which means that once a negative number is entered, ReLU will die. In this way, in the forward propagation process, it is not a problem. Some areas are sensitive and some are insensitive. But in the back propagation process, if you enter a negative number, the gradient will be completely zero, which has the same problem as the sigmoid function and tanh function.

- 死 ReLU 问题--当输入为负数时，ReLU 完全不工作，这意味着一旦输入负数，ReLU 就会死亡。这样，虽然在前向传播过程中，这不是问题了。有些区域敏感，有些区域不敏感。但在反向传播过程中，如果输入负数，梯度将完全为零，这与 sigmoid 函数和 tanh 函数有同样的问题。

2) We find that the output of the ReLU function is either 0 or a positive number, which means that the ReLU function is not a 0-centric function.

- ReLU 函数的输出要么是 0，要么是正数，这意味着 ReLU 函数不是一个以 0 为中心的函数。

### 4. Leaky ReLU Activation Function

Leaky Relu is an activation function specifically designed to compensate for the **dying ReLU problem.**

```python
def LeakyReLU(x):
    return np.where(x >= 0, x, 0.05 * x)
```

$f(y_i) = 
\begin{cases} 
y_i & ,\ \text{if } y_i > 0 \\
a_i y_i & ,\ \text{if } y_i \leq 0 
\end{cases}$

```python
# 生成x值
x = np.linspace(-10, 10, 400)
# 计算y值
y = LeakyReLU(x)
# 绘图
plt.plot(x, y)
# 添加网格
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
# 轴中心线
plt.axvline(x=0, color='k', linewidth=0.5) # 添加垂直Y=0轴线
# 标题和标签
plt.title("Function plot of LeakyReLu Function")
plt.xlabel("x")
plt.ylabel("y")
# 显示图像
plt.show()
```

#### Comparison between ReLU and Leaky ReLU Activation Functions


- The leaky ReLU adjusts the problem of zero gradients for negative value, by giving a very small linear component of x to negative inputs(0.01x).
  - ReLU 通过给负输入（0.01x）一个很小的 x 线性分量，解决了负值梯度为零的问题

- The leak helps to increase the range of the ReLU function. Usually, the value of a is 0.01 or so.
  - 泄漏有助于增大 ReLU 函数的范围。通常，a 的值为 0.01 左右。

- Range of the Leaky ReLU is (-infinity to infinity).

***Note :*** In theory, Leaky ReLU has all the advantages of ReLU, plus there will be no problems with Dead ReLU, but in actual operation, it has not been fully proved that Leaky ReLU is always better than ReLU. (Due to the complexity of actual problems and the characteristics of black box models)
  - 从理论上讲，Leaky ReLU 具有 ReLU 的所有优点，而且不会出现 Dead ReLU 的问题，但在实际运行中，Leaky ReLU 始终优于 ReLU 的事实尚未得到充分证明。(因为实际问题的复杂性以及黑盒模型的特性)

### 5. ELU (Exponential Linear Unit) Activation Function

ELU is also proposed to solve the problems of ReLU. In contrast to ReLUs, ELUs have negative values which pushes the mean of the activations closer to zero. Mean activations that are closer to zero enable faster learning as they bring the gradient closer to the natural gradient.

$$g(x) = \text{ELU}(x) =
\begin{cases}
x, & \text{if } x > 0 \\
\alpha (e^x - 1), & \text{if } x \leq 0
\end{cases}$$

```python
def ELU(x, alpha=1):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))
```

```python
# 生成x值
x = np.linspace(-10, 10, 400)
# 计算y值
y = ELU(x)
# 绘图
plt.plot(x, y)
# 添加网格
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
# 轴中心线
plt.axvline(x=0, color='k', linewidth=0.5) # 添加垂直Y=0轴线
# 标题和标签
plt.title("Function plot of ELU Function")
plt.xlabel("x")
plt.ylabel("y")
# 显示图像
plt.show()
```

#### Comparison between ReLU and ELU Activation Functions

Obviously, ELU has all the advantages of ReLU, and:

- No Dead ReLU issues,the mean of the output is close to 0, zero-centered.
    - 没有 ReLU 的问题，输出的平均值接近 0，以零为中心。

- In contrast to ReLUs, ELUs have negative values which allows them to push mean unit activations closer to zero like batch normalization but with lower computational complexity. Mean shifts toward zero speed up learning by bringing the normal gradient closer to the unit natural gradient because of a reduced bias shift effect.
    - 与 ReLUs 不同，ELUs 具有负值，这使得它们可以像批量归一化一样，将平均单元激活值推近零，但计算复杂度更低。由于偏移效应的减弱，平均值向零的移动会使正常梯度更接近单元自然梯度，从而加快学习速度。


- ELUs saturate to a negative value with smaller inputs and thereby decrease the forward propagated variation and information.
    - ELU 在输入较小的情况下饱和为负值，从而减少了前向传播的变化和信息。

One small problem is that it is slightly **more computationally intensive**. Similar to Leaky ReLU, although theoretically better than ReLU, there is currently no good evidence in practice that ELU is always better than ReLU.

一个小问题是，它的计算量稍大。与 Leaky ReLU 类似，虽然理论上 ELU 优于 ReLU，但目前还没有很好的实践证明 ELU 一定优于 ReLU。

### 6. PRelu (Parametric ReLU) Function

PReLU is also an improved version of ReLU.

![](https://miro.medium.com/v2/resize:fit:720/format:webp/1*BmVRlcbyW-ri03BRm9HV6g.png)

$$f(y_i) = 
\begin{cases}
y_i, & \text{if } y_i > 0 \\
a_i y_i, & \text{if } y_i \leq 0
\end{cases}$$

The parameter α is generally a number between 0 and 1, and it is generally relatively small, such as a few zeros.

- if aᵢ=0, f becomes ReLU

- if aᵢ>0, f becomes leaky ReLU

- if aᵢ is a learnable parameter, f becomes PReLU

#### Advantages of PReLU Activation Function

- In the negative region, PReLU has a small slope, which can also avoid the problem of ReLU death.
    - 在负区域，PReLU 的斜率较小，这也避免了 ReLU 死亡的问题。
- Compared to ELU, PReLU is a linear operation in the negative region. Although the slope is small, it does not tend to 0, which is a certain advantage.
    - 与 ELU 相比，PReLU 在负区域是线性运算。虽然斜率很小，但并不趋于 0，这也是它的一个优点。


### 7. Softmax Activation Function

Softmax is used as the activation function for ***multi-class classification problems*** where class membership is required on more than two class labels. For an arbitrary real vector of length K, Softmax can compress it into a real vector of length K with a value in the range (0, 1), and the sum of the elements in the vector is 1.

Softmax 用作多类分类问题的激活函数，在这种分类问题中，需要两个以上的类标签作为类成员。对于长度为 K 的任意实向量，Softmax 可以将其压缩为长度为 K 的实向量，其值范围为（0，1），且向量中的元素之和为 1。

![](https://miro.medium.com/v2/resize:fit:720/format:webp/1*t2FbP10uDFZrufDCYELWFA.jpeg)


Softmax is different from the normal max function: the max function only outputs the largest value, and Softmax ensures that smaller values have a smaller probability and will not be discarded directly. It is a "max" that is "soft"; it can be thought to be a probabilistic or "softer" version of the argmax function.

Softmax 与普通的 max 函数不同：max 函数只输出最大值，而 Softmax 则确保较小的值具有较小的概率，不会被直接丢弃。它是一种 "软 "的 "最大值"；可以认为它是 argmax 函数的概率或 "软 "版本。

The denominator of the Softmax function combines all factors of the original output value, which means that the different probabilities obtained by the Softmax function are related to each other.

Softmax 函数的分母结合了原始输出值的所有因素，这意味着 Softmax 函数得到的不同概率是相互关联的。

The major drawback in the softmax activation function is that it is -

- Non-differentiable at zero and ReLU is unbounded.
    - 零点无差别，ReLU 无边界。
- The gradients for negative input are zero, which means for activations in that region, the weights are not updated during backpropagation. This can create dead neurons that never get activated.
    - 负输入的梯度为零，这意味着对于该区域的激活，权重在反向传播过程中不会更新。这可能会产生永远无法激活的死神经元。

### 8. Swish Activation Function

The formula is: 
$$y = x * sigmoid(x)$$

Swish's design was inspired by the use of sigmoid functions for gating in LSTMs and highway networks. We use the same value for gating to simplify the gating mechanism, which is called self-gating.

Swish 的设计灵感来自 LSTM 和高速公路网络中使用 sigmoid 函数进行门控。我们使用相同的门控值来简化门控机制，这就是所谓的自门控。

![](https://miro.medium.com/v2/resize:fit:720/format:webp/1*avgXN89BZtU8Amfh7fex7w.png)


The advantage of self-gating is that it only requires a simple scalar input, while normal gating requires multiple scalar inputs. This feature enables self-gated activation functions such as Swish to easily replace activation functions that take a single scalar as input (such as ReLU) without changing the hidden capacity or number of parameters.

自门控的优势在于它只需要一个简单的标量输入，而普通门控需要多个标量输入。这一特点使得 Swish 等自门控激活函数可以在不改变隐藏容量或参数数量的情况下，轻松取代以单个标量作为输入的激活函数（如 ReLU）。

Note: Swish activation function can only be implemented when your neural network is ≥ 40 layers.

注：Swish 激活函数只能在神经网络层数≥ 40 层时使用。

The major advantages of the Swish activation function are as follows:

- Unboundedness is helpful to prevent the gradient from gradually approaching 0 during slow training, causing saturation.
    - 无边界性有助于防止梯度在缓慢训练过程中逐渐趋近于 0，从而导致饱和。

    - (At the same time, being bounded has advantages, because bounded active functions can have strong regularization, and larger negative inputs will be resolved.)
    - 同时，有界也有好处，因为有界的激活函数会有很强的正则化，较大的负输入也会被解决。

-  Derivative always > 0.
    - 导数始终大于 0。

-  Smoothness also plays an important role in optimization and generalization.
    - 平滑性在优化和泛化中也起着重要作用。

### 9. Maxout Activation Function

A Maxout layer is simply a layer where the activation function is the max of the inputs. As stated in the paper, even an MLP with 2 maxout units can approximate any function.

![](https://miro.medium.com/v2/resize:fit:640/format:webp/1*Ae28NCJH6ZYooc64jIPYQw.jpeg)



A single Maxout unit can be interpreted as making a piecewise linear approximation (PWL) to a real-valued function where the line segment between any two points on the graph of the function lies above or on the graph (convex function).

$$\text{ReLU} = \max(0, x), \quad \text{abs}(x) = \max(x, -x)$$


Maxout can also be implemented for a d-dimensional vector(V).

![](https://miro.medium.com/v2/resize:fit:720/format:webp/1*t8cNgQcRBYPtezzl0T1WVw.jpeg)

Consider, two convex functions h1(x) and h2(x), approximated by two Maxout units. By the above preposition, the function g(x) is a continuous PWL function.

Hence, it is found that a Maxout layer consisting of two Maxout units can approximate any continuous function arbitrarily well.

### 10. Softplus Activation Function

![](https://miro.medium.com/v2/resize:fit:720/format:webp/1*enmbqoxIDqtM6kODX7Sngg.png)

Softplus function: 
$$f(x) = ln(1+exp(x))$$

The derivative of softplus is 

$$f ′(x)=exp(x) / ( 1+exp⁡(x) ) = 1/ (1 +exp(−x ))$$

which is also called the logistic/sigmoid function对数/正比例函数.

The softplus function is similar to the ReLU function, but it is relatively smooth. It is unilateral suppression like ReLU.

- 软加函数与 ReLU 函数相似，但相对平滑。它与 ReLU 一样是单边抑制。

Generally speaking, these activation functions have their own advantages and disadvantages. All the good and bad must be obtained by experimenting them with various problem statements.

一般来说，这些激活函数各有利弊。所有的优缺点都必须通过在不同的问题陈述中进行实验来获得
